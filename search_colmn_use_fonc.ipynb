# Searching column data of dataset using function

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('forbes_mil.csv')
df.head()
```

```python
def fnd_sw(x):
    if x == "software":
        return x
    return np.NaN

def rnk(x):
    return x
def name(x):
    return x

def organization(x):
    x=str(x)#x variable turn to this "nan"
    if x == "nan":
        y="no organization"
        return y
    return x

sw_mil=df.agg({"rank":[rnk],"personName":[name],"source":[fnd_sw],"organization":[organization]})
sw_mil=sw_mil.dropna()
sw_mil.head(1)
```
